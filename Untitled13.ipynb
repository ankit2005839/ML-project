```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Load the Dataset
df = pd.read_csv('TASK2_dataset.csv')
print("Dataset shape:", df.shape)
# Sample of the dataset:
print(df.head())
```

```python
# 2. TF–IDF Vectorization for Each Keyword Column
def compute_tfidf(series):
    docs = series.tolist()
    N = len(docs)
    # Create a sorted vocabulary (list of unique tokens)
    vocab = sorted(set(docs))
    token_to_index = {token: i for i, token in enumerate(vocab)}
    # Build term-frequency (TF) matrix (each document is one token → one-hot)
    TF = np.zeros((N, len(vocab)))
    for i, token in enumerate(docs):
        TF[i, token_to_index[token]] = 1.0
    # Document frequency (number of documents in which a token appears)
    df_counts = np.sum(TF > 0, axis=0)
    # Compute Inverse Document Frequency (IDF)
    idf = {}
    for token, idx in token_to_index.items():
        idf[token] = np.log(N / (df_counts[idx] + 1e-9))
    # Compute TF-IDF matrix (multiplying each one-hot by idf weight)
    tfidf = np.zeros_like(TF)
    for token, idx in token_to_index.items():
        tfidf[:, idx] = TF[:, idx] * idf[token]
    return tfidf, vocab, idf
```

```python
tfidf1, vocab1, idf1 = compute_tfidf(df['keyword_1'])
tfidf2, vocab2, idf2 = compute_tfidf(df['keyword_2'])
tfidf3, vocab3, idf3 = compute_tfidf(df['keyword_3'])
```

```python
# 3. Dimensionality Reduction Using a Manual PCA
def pca_manual(X, n_components=2):
    # Center the data
    X_mean = np.mean(X, axis=0)
    X_centered = X - X_mean
    # Compute covariance matrix (features in columns)
    cov_matrix = np.cov(X_centered, rowvar=False)
    # Eigen-decomposition
    eigen_vals, eigen_vecs = np.linalg.eig(cov_matrix)
    # Sort the eigenvectors by descending eigenvalues
    sorted_indices = np.argsort(eigen_vals)[::-1]
    top_components = eigen_vecs[:, sorted_indices[:n_components]].real
    # Project the data onto the top components
    X_reduced = np.dot(X_centered, top_components)
    return X_reduced, X_mean, top_components

# Apply PCA individually to each TF-IDF matrix (reducing each to 2 dimensions)
pc1, mean1, comp1 = pca_manual(tfidf1, n_components=2)
pc2, mean2, comp2 = pca_manual(tfidf2, n_components=2)
pc3, mean3, comp3 = pca_manual(tfidf3, n_components=2)
```

```python
# 4. Combining the Three 2D Embeddings into One Final Embedding
# Here we take the average of the three 2D vectors for each song.
final_embedding = (pc1 + pc2 + pc3) / 3.0

# Plot the final embedding (PC1 vs PC2 — eigenvalue index 0 and 1)
plt.figure(figsize=(8,6))
plt.scatter(final_embedding[:,0], final_embedding[:,1], color='blue', alpha=0.6)
plt.xlabel('Eigenvalue index 0 (PC1)')
plt.ylabel('Eigenvalue index 1 (PC2)')
plt.title('Songs Projected onto PCA Space (PC1 vs PC2)')
plt.show()
```

```python
# 5. Clustering Using a Custom K-Means Implementation
def custom_kmeans(X, k, max_iter=100, tol=1e-4):
    np.random.seed(0)
    # Randomly initialize centroids from the existing data points
    indices = np.random.choice(len(X), k, replace=False)
    centroids = X[indices]
    for iteration in range(max_iter):
        # Compute Euclidean distances and assign clusters
        distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)
        # Update centroids as the mean of assigned points
        new_centroids = np.array([X[labels == i].mean(axis=0) if np.any(labels==i) 
                                  else centroids[i] for i in range(k)])
        if np.linalg.norm(new_centroids - centroids) < tol:
            break
        centroids = new_centroids
    return labels, centroids
```

```python
# 5a. Elbow Method: Plot Inertia vs. k
inertias = []
ks = range(2, 7)
for k in ks:
    labels_temp, centroids_temp = custom_kmeans(final_embedding, k)
    inertia = 0
    for i in range(k):
        cluster_points = final_embedding[labels_temp == i]
        inertia += np.sum(np.linalg.norm(cluster_points - centroids_temp[i], axis=1)**2)
    inertias.append(inertia)

plt.figure(figsize=(8,6))
plt.plot(list(ks), inertias, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia (Sum of Squared Distances)")
plt.title("Elbow Method for Optimal Cluster Count")
plt.show()
```

```python
# Choose a value for k based on the elbow plot (e.g., k=3)
k_final = 4
cluster_labels, centroids = custom_kmeans(final_embedding, k_final)
df['cluster'] = cluster_labels

# Plot the clustered songs (using different colors for each cluster)
plt.figure(figsize=(8,6))
colors = ['red', 'green', 'blue','yellow']
for i in range(k_final):
    points = final_embedding[cluster_labels == i]
    plt.scatter(points[:,0], points[:,1], color=colors[i], label=f'Cluster {i}', alpha=0.6)
plt.scatter(centroids[:,0], centroids[:,1], marker='X', s=200, color='black', label='Centroids')
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("K-Means Clustering on Combined Song Embedding")
plt.legend()
plt.show()

```

```python
# 6. Analysis: Cluster-Genre Distribution and Silhouette Score
distribution = df.groupby(['cluster', 'genre']).size().unstack(fill_value=0)
print("Cluster-Genre Distribution (counts):")
print(distribution)
percentage_distribution = distribution.div(distribution.sum(axis=1), axis=0) * 100
print("\nCluster-Genre Distribution (%):")
print(percentage_distribution.round(1))
```

```python
def silhouette_score(X, labels):
    n = len(X)
    unique_labels = np.unique(labels)
    total_score = 0
    for i in range(n):
        same_cluster = X[labels == labels[i]]
        a = np.mean(np.linalg.norm(X[i] - same_cluster, axis=1)) if len(same_cluster) > 1 else 0
        b = np.inf
        for lbl in unique_labels:
            if lbl == labels[i]:
                continue
            other_cluster = X[labels == lbl]
            if len(other_cluster) > 0:
                b = min(b, np.mean(np.linalg.norm(X[i] - other_cluster, axis=1)))
        if max(a, b) > 0:
            total_score += (b - a) / max(a, b)
    return total_score / n
```

```python
sil_score = silhouette_score(final_embedding, cluster_labels)
print("\nSilhouette Score: {:.2f}".format(sil_score))
```

```python
# 7. Genre Prediction for New Songs
# First, compute the majority (most frequent) genre for each cluster from training data.
cluster_majority = df.groupby('cluster')['genre'].agg(lambda x: x.value_counts().idxmax())
```

```python
# To predict the genre of a new song (given only its three keywords),
# we need to project each new token using the stored TF-IDF vocabularies and PCA parameters.
def project_new_token(token, vocab, idf, mean, comp):
    vec = np.zeros(len(vocab))
    if token in vocab:
        index = vocab.index(token)
        vec[index] = idf[token]  # Use the stored idf value for this token
    # Center the vector and project using PCA components
    projection = np.dot(vec - mean, comp)
    return projection
```

```python
def predict_genre(new_song):
    # new_song: list of three tokens [keyword_1, keyword_2, keyword_3]
    proj1 = project_new_token(new_song[0], vocab1, idf1, mean1, comp1)
    proj2 = project_new_token(new_song[1], vocab2, idf2, mean2, comp2)
    proj3 = project_new_token(new_song[2], vocab3, idf3, mean3, comp3)
    # Combine the projected vectors (average)
    final_proj = (proj1 + proj2 + proj3) / 3.0
    # Find the nearest cluster centroid
    distances = np.linalg.norm(centroids - final_proj, axis=1)
    cluster_assigned = np.argmin(distances)
    # Return the majority genre for that cluster
    return cluster_majority.loc[cluster_assigned]
```

```python
new_songs = [
    ['piano', 'calm', 'slow'],
    ['guitar', 'emotional', 'distorted'],
    ['synth', 'mellow', 'distorted']
]

for song in new_songs:
    genre_pred = predict_genre(song)
    print(f"Predicted genre for {song}: {genre_pred}")
```
